```python
#trovare i vini americani
df[df.country == "US"]
```

```python
#trovare i vini italiani
df[df.country == "Italy"]
```

```python
#condizioni: condizione and
#mostrare i vini italiani che costano più di 50 euro
df[(df.country == "Italy") & (df.price > 50)]
```

```python
#condizioni: condizione and
#mostrare i vini italiani che costano più di 550 euro
df[(df.country == "Italy") & (df.price > 550)]
```

```python
#vini italiani e francesi
#condizione or
df[(df.country == "Italy") | (df.country == "France")]
```

```python
#vini italiani, francesi e americani
#isin: è dentro 
df[df.country.isin(["Italy", "France", "US"])]
```

```python
#notnull: il valore esiste
df[df.price.notnull()]
```

```python
#isnull: il valore non esiste
df[df.price.isnull()]
#NaN: not a number
```

```python
#visualizzare tutti i prezzi dei vini
df["price"]
```

```python
#trovare min dei prezzi con il metodo min
min(df["price"])
```

```python
#trovare min con il metodo min di pandas(giusto)
df["price"].min()
```

```python
#trovare max con il metodo min di pandas(giusto)
df["price"].max()
```

```python
#trovare il vino con il prezzo massimo
df[df["price"] == df["price"].max()]["title"]
```

```python
#trovare il vino con il prezzo minimo
df[df["price"] == df["price"].min()]["title"]
```

```python
#trovare il nome del vino e il prezzo del vino con il prezzo massimo
df[df["price"] == df["price"].max()][["title", "price"]]
```

```python
#somma di tutti i vini
df["points"].sum()
```

```python
#media dei prezzi dei vini
round(df["price"].mean(), 2)
```

```python
#nomi del vino e della casa produttrice dei vini che costano più della media
df[df["price"] > df["price"].mean()][["winery", "title"]]
```

```python
#numero di vini americani
len(df[df["country"] == "US"])
```

```python
#numeri di vini americani con pandas
df[df["country"] == "US"]["country"].count()
#la seconda parentesi serve per specificare la colonna interessata
```
